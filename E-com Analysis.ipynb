```python
import calendar
```

```python
def count_weeks_in_month(date):
    _, num_days = calendar.monthrange(date.year, date.month)
#count number of days in month
    num_weeks = num_days/7
# count number of weeks in month
    return num_weeks

agg_clients_orders['weeks_in_month'] = agg_clients_orders['month'].apply(count_weeks_in_month)
```

Using Python, create RFM (Recency, Frequency, Monetary) segments of users to thoroughly assess your audience. In the clustering, you can choose the following metrics: R - time since the user's last purchase up to the current date, F - total number of purchases by the user over time, M - total monetary value of purchases over time. Describe in detail how you created the clusters. For each RFM segment, establish boundaries for the recency, frequency, and monetary metrics for interpreting these clusters. An example of such a description: RFM segment 132 (recency=1, frequency=3, monetary=2) has the following metric boundaries: recency ranges from 130 to 500 days, frequency ranges from 2 to 5 orders per week, monetary ranges from 1780 to 3560 rubles per week. 

```python
from datetime import timedelta
import matplotlib.pyplot as plt
```

```python
customers_orders_df=pd.merge(customers_df, orders_df, on = 'customer_id')
```

```python
# For each user, we set the earliest "approved_at" date from the selected stages as their last purchase date 
# to exclude delivery time and other factors
```

```python
NOW = customers_orders_items_df['order_approved_at'].max() + timedelta(days=1)
# Calculating the date for report generation (yesterday + 1 day = today).

NOW
```

```python
rfm = customers_orders_items_df \
    .groupby('customer_unique_id', as_index = False) \
    .agg({'order_approved_at': lambda x: (NOW - x.max()).days, \
          'product_id': lambda x: len(x), \
          'price': lambda x: x.sum()}) \
    .rename(columns={'order_approved_at': 'Recency', \
                     'product_id': 'Frequency', \
                     'price': 'Monetary'})

# Grouping by customers and calculating RFM as follows:
# R - Time in days since the user's last purchase up to the current date,
# F - Total number of purchases by the user over time,
# M - Total monetary value of purchases over time.
```

```python
rfm
```

```python
# We use quantiles for segmenting/clustering the dataset - dividing it into equal 20% portions. 
# This allows for a balanced segmentation where each segment comprises 20% of the data.
```

```python
quintiles = rfm.quantile([.2, .4, .6, .8]).to_dict()
quintiles
```

```python
# making 5 ranges:
# 1: 0-30 days
# 2: 31-60 days
# 3: 61-90 days
# 4: 91-180 days
# 5: 181-365 days
```

```python
def range_R(x):
    if x <= quintiles['Recency'][.2]:
        return 5
    elif x <= quintiles['Recency'][.4]:
        return 4
    elif x <= quintiles['Recency'][.6]:
        return 3
    elif x <= quintiles['Recency'][.8]:
        return 2
    else:
        return 1
# A smaller Recency value is better, so we assign a rank of 5 when prioritizing this metric.
def range_FM(x, c):
    if x <= quintiles[c][.2]:
        return 1
    elif x <= quintiles[c][.4]:
        return 2
    elif x <= quintiles[c][.6]:
        return 3
    elif x <= quintiles[c][.8]:
        return 4
    else:
        return 5    
# A higher Frequency and Monetary value is better, so we assign a rank of 5 when prioritizing these metrics.
```

```python
rfm['R'] = rfm['Recency'].apply(lambda x: range_R(x))
rfm['F'] = rfm['Frequency'].apply(lambda x: range_FM(x, 'Frequency'))
rfm['M'] = rfm['Monetary'].apply(lambda x: range_FM(x, 'Monetary'))
```

```python
# Combining the values into the RFM table and forming the 'RFM score'.
rfm['RFM Score'] = rfm['R'].map(str) + rfm['F'].map(str) + rfm['M'].map(str)
rfm.head()
```

```python
# The lower the R, the older the last purchase; the higher the R, the more recent the last purchase.
# The higher the F, the more frequently the customer made purchases.

segt_map = {
    r'[1-2][1-2]': 'sleeping non freq',
    r'[1-2][3-4]': 'sleeping avg freq',
    r'[1-2]5': 'sleeping high freq',
    r'[3-4]1': 'recent low freq buyers',
    r'[3-4]5': 'recent high freq buyers',
    r'[5]1': 'new low freq buyers',
    r'[5][2-4]': 'new avg freq buyers',
    r'[5]5': 'new high freq buyers'
}
rfm['Segment'] = rfm['R'].map(str) + rfm['F'].map(str)
rfm['Segment'] = rfm['Segment'].replace(segt_map, regex=True)
rfm.head()
```

```python
rfm_analysis=rfm.groupby('Segment', as_index=False)\
                .agg({'customer_unique_id':'count'})
rfm_analysis
# It's evident that there are 2305+4747+4557=11609 customers with frequent purchases compared to  16465+32877+32391=81733 with infrequent ones.
# Additionally, there's a category of "recent high-frequency buyers," and they could also be a focus since 
# they are customers with frequent purchases and relatively moderate Recency 
# (indicating they are still active and can be further nurtured and engaged).
```

```python
rfm_analysis_money=rfm.groupby('Segment', as_index=False)\
                                .agg({'Monetary':'mean'})\
                                .round()
rfm_analysis_money
# We can see that there are "whales" who make frequent purchases and remain active - 
# they bring the maximum sales to the service on a per-user basis.

# There are also dormant "whales" who made a lot of purchases in the past but are no longer active - 
# they are the second most profitable for the service. It would be beneficial to find ways to re-engage
# and motivate them to place orders.

# It's evident that repeat purchases are crucial for the service, and this is where they generate the maximum 
# revenue. Comparing in terms of revenue per customer, 
# low frequency versus high frequency is 225 versus 132 (almost twice as much).
```

```python

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

for i, p in enumerate(['R', 'F']):
    parameters = {'R':'Recency', 'F':'Frequency'}
    y = rfm[p].value_counts().sort_index()
    x = y.index
    ax = axes[i]
    bars = ax.bar(x, y, color='silver')
    ax.set_frame_on(False)
    ax.tick_params(left=False, labelleft=False, bottom=False)
    ax.set_title('Distribution of {}'.format(parameters[p]),
                fontsize=14)
    for bar in bars:
        value = bar.get_height()
        if value == y.max():
            bar.set_color('firebrick')
        ax.text(bar.get_x() + bar.get_width() / 2,
                value - 5,
                '{}\n({}%)'.format(int(value), int(value * 100 / y.sum())),
               ha='center',
               va='top',
               color='w')

plt.show()

# plot the distribution of M for RF score
fig, axes = plt.subplots(nrows=5, ncols=5,
                         sharex=False, sharey=True,
                         figsize=(10, 10))

r_range = range(1, 6)
f_range = range(1, 6)
for r in r_range:
    for f in f_range:
        y = rfm[(rfm['R'] == r) & (rfm['F'] == f)]['M'].value_counts().sort_index()
        x = y.index
        ax = axes[r - 1, f - 1]
        bars = ax.bar(x, y, color='silver')
        if r == 5:
            if f == 3:
                ax.set_xlabel('{}\nF'.format(f), va='top')
            else:
                ax.set_xlabel('{}\n'.format(f), va='top')
        if f == 1:
            if r == 3:
                ax.set_ylabel('R\n{}'.format(r))
            else:
                ax.set_ylabel(r)
        ax.set_frame_on(False)
        ax.tick_params(left=False, labelleft=False, bottom=False)
        ax.set_xticks(x)
        ax.set_xticklabels(x, fontsize=8)

        for bar in bars:
            value = bar.get_height()
            if value == y.max():
                bar.set_color('firebrick')
            ax.text(bar.get_x() + bar.get_width() / 2,
                    value,
                    int(value),
                    ha='center',
                    va='bottom',
                    color='k')
fig.suptitle('Distribution of M for each F and R',
             fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# It is evident that customers are evenly distributed over time in making purchases, with no dispersion. 
# However, the frequency is highly polarized - 87% of customers have a single order, and 12% have the maximum frequency of orders. 
# It can be seen that monetary distribution is even over recency, as the main part of orders is either fresh or very old.

# The absence of frequency 2-3-4 could be a growth point - to be found.
```
